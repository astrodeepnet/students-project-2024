```python
df1_numeric = df1.select_dtypes(include='number')

# Prendre un échantillon de 25% des données (Sinon MemoryError)
df1_sampled = df1_numeric.sample(frac=0.25)
```

```python
# Nombre de colonnes numériques
n_cols = len(df1_sampled.columns)

# Définir la taille de la figure et le layout
fig, axes = plt.subplots(n_cols, 1, figsize=(10, n_cols * 3))

# Si une seule colonne, on évite l'erreur en ajustant axes
if n_cols == 1:
    axes = [axes]

# Création d'un boxplot pour chaque colonne dans un subplot
#print sous forme de grid 

for i, col in enumerate(df1_sampled.columns):
    df1_sampled.boxplot(col, ax=axes[i], vert=False, patch_artist=True)
    axes[i].set_title(col)
    
plt.tight_layout()
plt.show()
```
