```python
# subset all chemical abundances (collumn name contain "_FE")
chemical_subset = df.filter(regex='_FE', axis=1) 

# merge with + df.filter(regex='_H', axis=1)
chemical_subset = pd.concat([chemical_subset, df.filter(regex='_H', axis=1)], axis=1)

# create a df without the chemical abundances
df_no_chemical = df.drop(chemical_subset.columns, axis=1)
```

```python
# GÃ©nÃ©rer les histogrammes
axes = df_drop.hist(figsize=(15, 12), bins=30, edgecolor='black')

# Si axes est une matrice (lorsque df_drop a beaucoup de colonnes), il faut itÃ©rer sur chaque axe
# Si axes est une liste d'axes 1D, on peut directement itÃ©rer dessus
if isinstance(axes, np.ndarray):
    axes = axes.flatten()  # Aplatir en une liste 1D si c'est un tableau 2D

# Appliquer semilogy Ã  chaque axe
for ax in axes:
    ax.set_yscale('log')  # Changer l'Ã©chelle de l'axe y en logarithmique

# Ajuster la mise en page pour Ã©viter les chevauchements
plt.tight_layout()

# Afficher les graphiques
plt.show()
```
