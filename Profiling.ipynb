```python
from astropy.io import fits
import pandas as pd
import matplotlib.pyplot as plt
```

```python

hdu1_data = hdulist[1].data
```

```python
df4 = pd.DataFrame(hdu1_data.tolist(), columns=hdu1_data.names) # TAKE 2/3 min
```

```python
# subset all chemical abundances (collumn name contain "_FE")
chemical_subset = df.filter(regex='_FE', axis=1) 

# merge with + df.filter(regex='_H', axis=1)
chemical_subset = pd.concat([chemical_subset, df.filter(regex='_H', axis=1)], axis=1)

# create a df without the chemical abundances
df_no_chemical = df.drop(chemical_subset.columns, axis=1)
```

```python
df = df_no_chemical
```

```python

# remove all _ERR and _SPEC columns
df = df.drop(df.filter(regex='_ERR', axis=1), axis=1)
df = df.drop(df.filter(regex='_SPEC', axis=1), axis=1)
```

```python
missing_values = df.isnull().sum()

chemical_subset_missing_values = chemical_subset.isnull().sum()
```

```python
missing_values = missing_values[missing_values > 0]    

missing_values.sort_values(ascending=False, inplace=True)

chemical_subset_missing_values = chemical_subset_missing_values[chemical_subset_missing_values > 0]

chemical_subset_missing_values.sort_values(ascending=False, inplace=True)


```

```python
if len(missing_values) > 0:

    plt.figure(figsize=(15, 8))  
    missing_values.plot(kind='bar')

    plt.title('Nombre de valeurs nulles par colonne')
    plt.xlabel('Nom des colonnes')
    plt.ylabel('Nombre de valeurs nulles')
    plt.xticks( ha='right', fontsize=8)  
    


    plt.tight_layout()  # Assure que tout s'affiche bien dans la figure
    plt.show()
```

```python
# Display collumns contained in df
display(df.columns)

# for all collumns in df display missing values number
for col in df.columns:
    print(f"{col} : {df[col].isnull().sum()}")
```

```python
# correlation matrix on df

corr = df.corr()
plt.figure(figsize=(15, 12))
plt.matshow(corr, fignum=1)
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.show()
```

```python
if len(chemical_subset_missing_values) > 0:

    plt.figure(figsize=(15, 8))  
    chemical_subset_missing_values.plot(kind='bar')

    plt.title('Nombre de valeurs nulles par colonne')
    plt.xlabel('Nom des colonnes')
    plt.ylabel('Nombre de valeurs nulles')
    plt.xticks( ha='right', fontsize=8)  


    plt.tight_layout()  # Assure que tout s'affiche bien dans la figure
    plt.show()
```

```python
low_variability_columns = []

for col in df.columns:
    try:
        # Si la colonne contient moins de 5 valeurs uniques, on l'ajoute Ã  la liste
        if df[col].nunique() < 5:
            low_variability_columns.append(col)
    except TypeError:
        print(f"error with column {col}")

print("Colonnes avec faible variabilitÃ© :", low_variability_columns)

```

```python
# create an error dataset containing all errors 
chemical_errors_subset = chemical_subset.filter(regex='_ERR', axis=1)

# create an error dataset containing all specs
chemical_specs_subset = chemical_subset.filter(regex='_SPEC', axis=1)

# create a dataset containing all the flags
flags_subset = chemical_subset.filter(regex='FLAG', axis=1)

# remove flags, err and spec from the chemical_subset
chemical_subset = chemical_subset.drop(chemical_errors_subset.columns, axis=1)
chemical_subset = chemical_subset.drop(chemical_specs_subset.columns, axis=1)
chemical_subset = chemical_subset.drop(flags_subset.columns, axis=1)




```
