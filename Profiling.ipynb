```python
# subset all chemical abundances (collumn name contain "_FE")
chemical_subset = df.filter(regex='_FE', axis=1) 

# merge with + df.filter(regex='_H', axis=1)
chemical_subset = pd.concat([chemical_subset, df.filter(regex='_H', axis=1)], axis=1)

# create a df without the chemical abundances
df_no_chemical = df.drop(chemical_subset.columns, axis=1)
```

```python
df = df_no_chemical
```

```python
missing_values = df.isnull().sum()

chemical_subset_missing_values = chemical_subset.isnull().sum()
```

```python
missing_values = missing_values[missing_values > 0]    

missing_values.sort_values(ascending=False, inplace=True)

chemical_subset_missing_values = chemical_subset_missing_values[chemical_subset_missing_values > 0]

chemical_subset_missing_values.sort_values(ascending=False, inplace=True)


```

```python
if len(missing_values) > 0:

    plt.figure(figsize=(15, 8))  
    missing_values.plot(kind='bar')

    plt.title('Nombre de valeurs nulles par colonne')
    plt.xlabel('Nom des colonnes')
    plt.ylabel('Nombre de valeurs nulles')
    plt.xticks( ha='right', fontsize=8)  
    


    plt.tight_layout()  # Assure que tout s'affiche bien dans la figure
    plt.show()
```

```python
if len(chemical_subset_missing_values) > 0:

    plt.figure(figsize=(15, 8))  
    chemical_subset_missing_values.plot(kind='bar')

    plt.title('Nombre de valeurs nulles par colonne')
    plt.xlabel('Nom des colonnes')
    plt.ylabel('Nombre de valeurs nulles')
    plt.xticks( ha='right', fontsize=8)  


    plt.tight_layout()  # Assure que tout s'affiche bien dans la figure
    plt.show()
```

```python
low_variability_columns = []

for col in df.columns:
    try:
        # Si la colonne contient moins de 5 valeurs uniques, on l'ajoute Ã  la liste
        if df[col].nunique() < 5:
            low_variability_columns.append(col)
    except TypeError:
        print(f"error with column {col}")

print("Colonnes avec faible variabilitÃ© :", low_variability_columns)

```
